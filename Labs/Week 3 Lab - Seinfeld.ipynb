### Install SSH
```
sudo apt-get install ssh

#check
ssh localhost

```

### Secure SSH
Disable Password Authentication and Root Login
```
sudo nano /etc/ssh/sshd_config
```
PasswordAuthentication no
PermitRootLogin no

Restart SSH (maybe)
```
/etc/init.d/ssh restart
```

### Enable Firewall
```
sudo ufw enable
```

### Update Firefox
```
sudo apt-get update
sudo apt-get install firefox
```

### Unlock User
```
cd /etc/
sudo usermod -U USERNAME
```

check 
```
sudo cat shadow
```

### Change Other Users Password
```
sudo passwd USERNAME
```

### New Root Password/Disable Password
```
sudo passwd
sudo passwd -l root
```

### Groups
- List all groups
```
cat /etc/group
```

### Enforce Password Requirement
- /etc/login.defs
- PASS_MIN_DAYS 7
- PASS_MAX_DAYS 90
- PASS_WARN_AGE 14

### Add Minimum Password Length
- Open /etc/pam.d/common-password
- Add minlen-8 to the end

### Implement Account Lockout Policy
- Open /etc/pam.d/common-auth
- Add deny=5 unlock_time=1800 to the end of line with pam_tally2.so

### Add User to Admin Group
```
sudo adduser USERNAME sudo
```
