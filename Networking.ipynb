# Networking

Computer networks are characterized by their size and their purpose. The size can be based on geographic areas or on the number of computers that are part of the network. Networks can cover from a single device in a room to a million devices around the world.

11 different types of Networking

## Networks Based on Size
### WAN (World Area Network)
- spans cities, states, countries
- runs at a slower speed
- ex: internet

### MAN (Metropolitan Area Network)
- city wide network (also college campus or smaller region)
- can cover an area from several miles to ten miles
- often used to connect several LANs together to form a big network
- ex: telcomms, internet access, businesses

### CAN (Campus Area Network)
- typically designed for a college campus
- can be referred to a MAN


### LAN (Local Area Network)
- computer network at a single site
- provides small area network, where people need to share resources amongst themselves but not with the rest of the world.
- mostly relies on wired connections for increased speed and security, but wireless connections can be applicable.
- typically in a home or office building where need to access files on a central server or print a document to one or more central printer.

### PAN (Personal Area Network)
- computer organized around an individual within a single building
- can include one or more device
    - computer
    - telephones
    - video game consoles
    - peripheral devices (ex: printer, keyboard, camera, TV)
- Allows you to:
    - Send a document to the printer upstairs while at your laptop downstairs
    - Upload a photo from your phone to your desktop
    - Watch movies through online streaming service to your TV
- Pros & Cons
    - Has good flexibility
    - Bluetooth have safe, but slow data transfer speed

### WLAN (Wireless Local Area Network)
- points to small geographic environment
- access point and a client
- Hi-Frequency Radio Waves

# Topologies
How the network would be laid out.

Bus star ring mesh tree

Bus
- network devices connected to a single cable
- not meant for large networks

Star
- central hub/switch that all devices are connected to

Ring
- Each workstation connected to two other groups (adjacent neighbors)

Mesh
- Every node connected to every other node in a network
