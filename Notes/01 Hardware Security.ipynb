# Hardware Security

## BIOS and UEFI
**Basic Input/Output System (System BIOS, ROM BIOS)**
- Software/Firmware used to start your computer
- POST - Power-On Self-Test
- BIOS settings are stored in ono volatile BIOS memory
    - Formerly CMOS
    - Now uses flash memory
    
**Legacy BIOS**
- Original BIOS
- No drivers for modern network, video, and storage devices

**UEFI BIOS**
- Unified Extensible Firmware Interface
- BIOS version for MACs
- Based on Intel's EFI (Extensible Firmware Interface)
- Designed to replace the legacy BIOS
- Boot from large (> 2.2 TB) GUID partition table (GPT) disks

**Secure Boot**
- Part of the UEFI specification
- Digitally sign known-good software
    - Software won't run without the proper signature
- Support in many different operating systems
    - Windows 8, 8.1, Windows Server 2012 and 2012R
    - Linux Fedora, openSUSE, Ubuntu Built into the BIOS

**BIOS Configurations**
- Use BIOS Configuration key during startup
- Configure and enable/disable components
- May include built-in hardware diagnostics
- [TIP] Always have a backup before making changes

**The "CMOS" Battery**
- Not needed for today's flash storage
- Maintains older BIOS configurations
- May only be used to maintain date/time
- Having a bad battery will require a BIOS configuration on every boot
- Can reset the BIOS configuration by removing the battery on older systems

**BIOS Passwords**
- 2 Levels of Password
     1. User Password - Needed to start the computer
     2.Supervisor Password - Needed for configuration changes
     
**LoJack for Laptops**
- Built into the BIOS
    - Software installed into the OS
    - Will reinstall itself if removed or new storage drive installed
- "Phone home" function provides location information
- Theft mode: remotely locks the laptop and/or delete files and can force a startup password

**Upgrading your BIOS firmware**
- Identify current firmware version
- Read the upgrade documentation
- Have a reliable power source

**Trusted Platform Module (TPM)**
- Adds advanced cryptographic funcitons
- Used by full disk encryption software
- May be a separate module or integrated into the motherboard



BIOS-Level Password


Locking down boot order in BIOS
1. Drive Encryption
